```python
dataset = dataset[dataset['original_language'] == 'en']
```

```python
import ast

def clean_genres(text):
    genres = []
    
    try:
        text = ast.literal_eval(text)
        
        for item in text:
            genres.append(item['name'])
            
        return " ".join(genres)
    
    except:
        return ""
```

```python
dataset['genres'] = dataset['genres'].apply(clean_genres)
dataset['genres']
```

```python
dataset.rename(columns={'original_title':'Title'},inplace=True)
dataset
```

```python
dataset=dataset[['Title','genres']]
dataset
```

```python
dataset.genres
```

```python
dataset.isnull().sum()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfi=TfidfVectorizer(stop_words='english')
tfi_vector=tfi.fit_transform(dataset['genres'])
tfi_vector
```

```python
tfi_vector.shape
```

```python
tfi_vector.toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar=cosine_similarity(tfi_vector)
```

```python
df=pd.DataFrame(similar,columns=dataset.Title,index=dataset.Title)
df
```

```python
df.columns.name=None
```

```python
df.index.name=None
```

```python
df
```

```python
df.columns
```

```python
def movie_recommendation(movie_name,top_rec):
    if movie_name not in df.index:
        return f"{movie_name} not found in the dataset"
    similarity=df[movie_name].sort_values(ascending=False)
    result=similarity.iloc[1:top_rec+1]
    result=round(result,2)*100
    return result.to_frame("Similarity")
```

```python
movie_recommendation('Father of the Bride Part II',5)
```
